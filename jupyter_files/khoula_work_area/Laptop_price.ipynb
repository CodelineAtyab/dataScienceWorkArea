## Feature Extraction

```python
# Import the regular expression module
import re

def extract_ram_size(ram_description):
    try:
        ram_size = int(ram_description.replace('GB', '').strip())
        return ram_size
    except ValueError:
        return None
def extract_ScreenResolution(screen_description):
    try:
        ScreenResolution = screen_description.split()[-1]
        return ScreenResolution
    except ValueError:
        return None

def extract_cpu(cpu):
    ghz_match = re.search(r'(\d+(?:\.\d+)?+)GHz', cpu)
    if ghz_match:
        return float(ghz_match.group(1))
    else:
        return None
def extract_cpu_com(cpu):
    cpu_company = cpu.split()[0]
    if cpu_company:
        return cpu_company
    else:
        return None
def extract_Memory_size(memory_txt):
    memory = 0
    memory_matchTB = re.search(r'(\d+(?:\.\d+)?+)TB', memory_txt)
    if memory_matchTB:
        memory =  int(float(memory_matchTB.group(1)) * 1024)
    memory_matchGB = re.search(r'(\d+(?:\.\d+)?+)GB', memory_txt)
    if memory_matchGB:
        memory= memory+ int(memory_matchGB.group(1))
    return memory
def extract_Memory_type(memory_txt):
    memory_txt = re.sub(r'\w*\d\w*', '', memory_txt).strip()
    memory_txt = re.sub(r'\.', '', memory_txt).strip()    
    return memory_txt
def extract_Touchscreen(screen):
    if screen.includes('Touchscreen'):
        return 1
    else:
        return 0
```

```python
# Extract RAM size in GB from "Ram" column
df_train['Ram'] = df_train['Ram'].apply(extract_ram_size)
df_test['Ram'] = df_test['Ram'].apply(extract_ram_size)
```

```python
df_train['ScreenResolution'] = df_train['ScreenResolution'].apply(extract_ScreenResolution)
df_test['ScreenResolution'] = df_test['ScreenResolution'].apply(extract_ScreenResolution)
```

```python
# Extract GHz value from "Cpu" column and create a new column "CpuGHz"
df_train['CpuGHz'] = df_train['Cpu'].apply(extract_cpu)
df_train['CpuCompany'] = df_train['Cpu'].apply(extract_cpu_com)
df_test['CpuGHz'] = df_test['Cpu'].apply(extract_cpu)
df_test['CpuCompany'] = df_test['Cpu'].apply(extract_cpu_com)
```

```python
df_train.drop(columns=['Cpu'], inplace = True)
df_test.drop(columns=['Cpu'], inplace = True)
df_train
```

```python
df_train['Memory'].values[:10]
```

```python
df_train['MemorySize'] = df_train['Memory'].apply(extract_Memory_size)
df_test['MemorySize'] = df_test['Memory'].apply(extract_Memory_size)
```

```python
df_train['MemoryType'] = df_train['Memory'].apply(extract_Memory_type)
df_test['MemoryType'] = df_test['Memory'].apply(extract_Memory_type)
```

```python
df_train.drop(columns=['Memory'], inplace = True)
df_test.drop(columns=['Memory'], inplace = True)
```

```python
df_train["Weight"] = df_train["Weight"].str.replace("kg", "").astype(float)
df_test["Weight"] = df_test["Weight"].str.replace("kg", "").astype(float)
```
