```python
import pandas as pd
import numpy as np
from random import randrange
from pandas import Series
from datetime import datetime
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.ar_model import AutoReg
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.tsa.api import AutoReg
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.statespace.sarimax import SARIMAX 

from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('../../dataset/AirPassengers.csv')
df
```

```python
# Make sure your "Date" column is in datetime format
df['Month'] = pd.to_datetime(df['Month'])
 
# Sorting the data by date (if not sorted)
df = df.sort_values(by='Month')
 
# Resetting the index
df.set_index('Month', inplace=True)
 
df.dropna(inplace=True)
df
```

```python
result = seasonal_decompose(df, model='additive', period=12)
plt.figure(figsize=(20, 20), dpi=100)

result.plot()
plt.show()
```

```python
def plot_series(series):
    plt.figure(figsize=(12,6))
    plt.plot(series, color='red')
    plt.ylabel('# Air Passengers', fontsize=16)

    for year in range(1949, 1963):
        plt.axvline(datetime(year,1,1), linestyle='--', color='k', alpha=0.5)
```

```python
plot_series(df)
```

## Normalize

```python
avg, dev = df.mean(), df.std()
```

```python
passengers = (df - avg) / dev
```

```python
plot_series(passengers)
plt.axhline(0, linestyle='--', color='k', alpha=0.3)
```

## Take First Difference to Remove Trend

```python
passengers = passengers.diff().dropna()
```

```python
plot_series(passengers)
plt.axhline(0, linestyle='--', color='k', alpha=0.3)
```

## Remove Increasing Volatility


```python
annual_volatility = passengers.groupby(passengers.index.year).std()
```

```python
annual_volatility
```

```python
passengers_annual_vol = passengers.index.map(lambda d: float(annual_volatility.loc[d.year]))
passengers_annual_vol
```

```python
passengers = passengers['#Passengers'] / passengers_annual_vol
```

```python
passengers
```

```python
plot_series(passengers)
plt.axhline(0, linestyle='--', color='k', alpha=0.3)
```

## Remove Seasonality


```python
month_avgs = passengers.groupby(passengers.index.month).mean()
```

```python
month_avgs
```

```python
passengers_month_avg = passengers.index.map(lambda d: month_avgs.loc[d.month])
```

```python
passengers_month_avg
```

```python
passengers = passengers - passengers_month_avg
```

```python
plot_series(passengers)
plt.axhline(0, linestyle='--', color='k', alpha=0.3)
```

```python

def undo_transformations(predictions, mu, sigma, month_avgs, annual_volatility):
    # Initialize the list to store original predictions
    orig_predictions = []
    
    # Recover the first prediction
    first_pred = (predictions[0] + month_avgs.loc[predictions.index.month[0]]) * annual_volatility.loc[predictions.index.year[0]] * sigma + mu
    orig_predictions.append(first_pred)
    
    # Iterate through the predictions to recover the original series
    for i in range(1, len(predictions)):
        # Recover the current z-score
        monthly_avg = month_avgs.loc[predictions.index.month[i]]
        sigma_year = annual_volatility.loc[predictions.index.year[i]]
        
        # Calculate the adjusted z-score
        adjusted_z_score = (predictions[i] + monthly_avg) * sigma_year
        
        # Recover the original z-score
        current_z_score = adjusted_z_score + orig_predictions[-1] - mu
        
        # Recover the original value
        current_value = current_z_score * sigma + mu
        orig_predictions.append(current_value)
    
    return np.array(orig_predictions).flatten()
```

## Build the models

```python
# Drop rows with NaN values resulting from creating lag features
df.dropna(inplace=True)

# Split the data into training and 2 years testing sets
train_data = passengers[:-24]
test_data = passengers[-24:]

```

```python
test_data
```

```python
lag_order = 30 # Adjust this based on the ACF plot
ar_model = AutoReg(train_data, lags=lag_order)
ar_results = ar_model.fit()
```

```python
# Make predictions on the test set
y_pred = ar_results.predict(start=len(train_data), end=len(train_data) + len(test_data) - 1, dynamic=False)
#print(y_pred)

# Calculate MAE and RMSE
mae = mean_absolute_error(test_data, y_pred)
rmse = np.sqrt(mean_squared_error(test_data, y_pred))
r2 = r2_score(test_data, y_pred)
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'R Squared value: {r2:.2f}')
```

```python
y_test = undo_transformations(test_data, avg, dev, month_avgs, annual_volatility)
y_pred_t = undo_transformations(y_pred, avg, dev, month_avgs, annual_volatility)
```

```python
# Calculate MAE and RMSE
mae = mean_absolute_error(y_test, y_pred_t)
rmse = np.sqrt(mean_squared_error(y_test, y_pred_t))
r2 = r2_score(y_test, y_pred_t)
print(f'Mean Absolute Error: {mae:.2f}')
print(f'Root Mean Squared Error: {rmse:.2f}')
print(f'R Squared value: {r2:.2f}')
```
