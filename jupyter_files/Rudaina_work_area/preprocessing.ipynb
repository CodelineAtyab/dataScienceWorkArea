```python
# Function to split Cpu column into CPUModel and CPUClockSpeed
def split_cpu_info(cpu):
    parts = cpu.split()
    
    cpu_model = ' '.join(parts[:len(parts)-1])
    cpu_clock_speed = parts[-1]
    
    return pd.Series([cpu_model, cpu_clock_speed], index=['CPUModel', 'CPUClockSpeed'])
```

```python
# Apply the function to split Cpu column and assign to new columns
df1[['CPUModel', 'CPUClockSpeed']] = df1['Cpu'].apply(split_cpu_info)
```

```python
df1
```

```python
df1 = df1.drop('Cpu', axis=1)
```

```python
df1.isnull().sum()
```

```python
# Clean the 'CPUClockSpeed' column and convert to float
df1['CPUClockSpeed'] = df1['CPUClockSpeed'].str.replace('GHz', '')
df1['CPUClockSpeed'] = pd.to_numeric(df1['Ram'], errors='coerce')
```

```python
df1.isnull().sum()
```

```python
# Clean the 'Ram' column and convert to float
df1['Ram'] = df1['Ram'].str.replace('GB', '')
df1['Ram'] = pd.to_numeric(df1['Ram'], errors='coerce')
```

```python
df1.isnull().sum()
```

```python
df1
```

```python
# Function to convert Memory sizes to numerical values in GB
def convert_memory_size(memory):
    if 'GB' in memory:
        # Extract numeric part and convert to float
        numeric_part = memory.split('GB')[0].strip()
        numeric_value = float(numeric_part)
        return numeric_value
    elif 'TB' in memory:
        # Convert TB to GB (1 TB = 1024 GB)
        numeric_part = memory.split('TB')[0].strip()
        numeric_value = float(numeric_part) * 1024
        return numeric_value
    else:
        return None  # Handle cases where format doesn't match
```

```python
# Apply the function to convert Memory sizes to numerical values in GB
df1['Memory_Size'] = df1['Memory_Size'].apply(convert_memory_size)
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
df1['Gpu'].unique()
```

```python
# Function to extract GPU company from Gpu column
def extract_gpu_company(gpu):
    return gpu.split()[0]

# Apply the function to extract GPU company and create a new column
df1['GPU_Company'] = df1['Gpu'].apply(extract_gpu_company)
```

```python
df1
```

```python
# Function to process ScreenResolution column
def process_screen_resolution(resolution):
    parts = resolution.split()
    
    # Check if there are additional descriptors
    if len(parts) > 1:
        screen_resolution = parts[-1]  # Last part is the resolution
        screen_type = ' '.join(parts[:-1])  # Join remaining parts as Screen_Type
    else:
        screen_resolution = resolution
        width, height = map(int, screen_resolution.split('x'))
        if width >= 1920 and height >= 1080:
            screen_type = 'Full HD'
        elif width >= 1600 and height >= 900:
            screen_type = 'HD+'
        else:
            screen_type = 'Other'  # Adjust as per your specific criteria
    
    return pd.Series([screen_resolution, screen_type], index=['ScreenResolution', 'Screen_Type'])
```

```python
# Apply the function and create new columns
df1[['ScreenResolution', 'Screen_Type']] = df['ScreenResolution'].apply(process_screen_resolution)
```
