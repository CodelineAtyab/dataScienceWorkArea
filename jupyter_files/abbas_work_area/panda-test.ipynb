```python
df['Price'] = pd.to_numeric(df['Price'].astype(str).str.replace(',', ''), errors='coerce')

# Calculate the mean price for Samsung
companies = df['company'].unique()
mean_prices = {}
for company in companies:
    mean_prices[company] = df[df['company'] == company]['Price'].mean()

company_names = list(mean_prices.keys())
mean_price_values = list(mean_prices.values())
plt.figure(figsize=(25,15))
plt.bar(company_names, mean_price_values)

plt.xlabel('Company', fontsize=16)
plt.ylabel('Mean Price', fontsize=16)
plt.title('Mean Prices', fontsize=25)

plt.show()
```

```python
category = df.select_dtypes(exclude = np.number).columns
category
```

```python
df1 = df.copy()
classes = dict()
le = LabelEncoder()
for col in category:
    df1[col]= le.fit_transform(df1[col])
    classes[col] = le.classes_
df3 = df1.copy()
df1
```

```python
X = df1.drop(columns = ['Price', 'Unnamed: 0']).values
X
```

```python
y = df1['Price'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test.shape
```

```python
reg = LinearRegression() 
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
np.mean(np.square(y_test - y_pred))
```

```python
np.mean(np.absolute(y_test - y_pred))
```

```python
min(y), max(y)
```

```python
y_mean = np.mean(y_test)
TSS = np.sum(np.square(y_test - y_mean))
RSS = np.sum(np.square(y_test - y_pred))
1- RSS/ TSS
```

```python

r2_score(y_test, y_pred), mean_squared_error(y_test, y_pred), mean_absolute_error(y_test, y_pred)
```
