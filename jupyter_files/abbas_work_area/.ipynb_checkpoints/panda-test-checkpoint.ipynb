```python
df.boxplot("Rating")
Q0 = df['Rating'].quantile(0)
Q1 = df['Rating'].quantile(0.25)
Q2 = df['Rating'].quantile(0.5)
Q3 = df['Rating'].quantile(0.75)
Q4 = df['Rating'].quantile(1)
plt.text(1.1, Q0, 'Q0: {:.2f}'.format(Q0), horizontalalignment='left', verticalalignment='center', color='grey')
plt.text(1.1, Q1, 'Q1: {:.2f}'.format(Q1), horizontalalignment='left', verticalalignment='center', color='blue')
plt.text(1.1, Q2, 'Q2 (Median): {:.2f}'.format(Q2), horizontalalignment='left', verticalalignment='center', color='green')
plt.text(1.1, Q3, 'Q3: {:.2f}'.format(Q3), horizontalalignment='left', verticalalignment='center', color='red')
plt.text(1.1, Q4, 'Q4: {:.2f}'.format(Q4), horizontalalignment='left', verticalalignment='center', color='orange')
```
