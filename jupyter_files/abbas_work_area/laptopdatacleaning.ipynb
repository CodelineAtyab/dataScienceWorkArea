```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv(r"../../dataset/laptopData.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method = 'bfill')
df
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
df1['Ram'] = df1['Ram'].str.replace("GB", "")
df1
```

```python
df1['Weight'] = df1['Weight'].str.replace("kg", "")
df1
```

```python
def split_memory_column(df1, column_name):
    # Split the column into two new columns
    df1[['Memory', 'Type']] = df1[column_name].str.split(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_memory_column(df1, 'Memory')
df1
```

```python
df1['Memory'] = df1['Memory'].str.replace("GB", "")
df1
```

```python
def split_GPU_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['Gpu', 'GPU_Ghz']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_GPU_column(df1, 'Gpu')
df1
```

```python
def split_Screen_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['ScreenResolution', 'Resolution']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_Screen_column(df1, 'ScreenResolution')
df1
```

```python
def split_Cpu_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['Cpu', 'Cpu_Ghz']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_Cpu_column(df1, 'Cpu')
df1
```

```python
category = df1.select_dtypes(exclude = np.number).columns
le = LabelEncoder()
classes = dict()
indices_to_print = [3, 6, 7, 10, 14, 15]
cat = df1.columns[indices_to_print]
for i in category:
    df1[i] = le.fit_transform(df1[i])
    classes[i] = le.classes_
df1
```

```python
X = df1.drop(columns = ['Price', 'Unnamed: 0']).values
X
```

```python
y = df1['Price'].values
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_test.shape
```

```python
reg = LinearRegression() 
reg.fit(X_train, y_train)
```

```python
y_pred = reg.predict(X_test)
```

```python
y_pred[:10]
```

```python
y_test[:10]
```

```python
np.mean(np.square(y_test - y_pred))
```

```python
np.mean(np.absolute(y_test - y_pred))
```

```python
min(y), max(y)
```

```python
y_mean = np.mean(y_test)
TSS = np.sum(np.square(y_test - y_mean))
RSS = np.sum(np.square(y_test - y_pred))
1- RSS/ TSS
```
