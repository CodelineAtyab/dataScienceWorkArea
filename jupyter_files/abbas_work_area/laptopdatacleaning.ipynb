```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
```

```python
df = pd.read_csv(r"../../dataset/laptopData.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.fillna(method = 'bfill')
df
```

```python
df.isnull().sum()
```

```python
df1 = df.copy()
df1['Ram'] = df1['Ram'].str.replace("GB", "")
df1
```

```python
df1['Weight'] = df1['Weight'].str.replace("kg", "")
df1
```

```python
def split_memory_column(df1, column_name):
    # Split the column into two new columns
    df1[['Memory', 'Type']] = df1[column_name].str.split(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_memory_column(df1, 'Memory')
df1
```

```python
df1['Memory'] = df1['Memory'].str.replace("GB", "")
df1
```

```python
def split_GPU_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['Gpu', 'GPU_Ghz']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_GPU_column(df1, 'Gpu')
df1
```

```python
def split_Screen_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['ScreenResolution', 'Resolution']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_Screen_column(df1, 'ScreenResolution')
df1
```

```python
def split_Cpu_column(df1, column_name):
    # Split the column into two new columns based on the last space
    df1[['Cpu', 'Cpu_Ghz']] = df1[column_name].str.rsplit(' ', n=1, expand=True)
    return df1
```

```python
df1 = split_Cpu_column(df1, 'Cpu')
df1
```

```python
le = LabelEncoder()
classes = dict()
indices_to_print = [3, 6, 7, 10, 14, 15]
cat = df1.columns[indices_to_print]
for i in cat:
    df1[i] = le.fit_transform(df1[i])
    classes[i] = le.classes_
df1
```
