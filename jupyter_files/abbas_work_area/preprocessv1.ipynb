```python
def age(Customer_Age):
    if Customer_Age<25:
        return "Youth (<25)"
    if Customer_Age<35:
        return "Young Adults (25-34)"
    return "Adults (35-64)"
```

```python
df.Age_Group[df.Age_Group.isnull()] = age(int(df.Customer_Age[df.Age_Group.isnull()]))
```

```python
df.iloc[15]
```

```python
df['Age_Group'] = df['Customer_Age'].apply(age)
```
