Image Processing

```python
import cv2
import matplotlib.pyplot as plt
```

```python
#save img and read rgb img
img = cv2.imread('dogg.jpg')
plt.imshow(img)
```

```python
#output img with window name as 'image'
img1 = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img1)
```

```python
img2 = cv2.imread('moon.bmp')
img3 = (img2 +20)%255
plt.imshow(img3)
```

```python
import numpy as np
plt.imshow(np.zeros((20, 20, 3))+255)
```

```python
grey_img = cv2.cvtColor(img1, cv2.COLOR_RGB2GRAY)
plt.imshow(grey_img, cmap='grey')
```

```python
grey = grey_img.T
plt.imshow(grey, cmap='grey')
```

```python
#Scaling

img4 = cv2.imread("fruits_detailfeature.jpg")
r,c = img4.shape[:2]
print('r,c = ', r,c)
resized_img = cv2.resize(img4, (2*r, 2*c), interpolation=cv2.INTER_CUBIC)
r,c = resized_img.shape[:2]
print ('r,c = ', r,c)
cv2.imwrite("resized_fruit.jpg", resized_img)
plt.imshow(resized_img, cmap='grey')
plt.xticks([]), plt.yticks([])
plt.title("Using plt on img read by openCV")
plt.show()
cv2.namedWindow('original img', cv2.WINDOW_NORMAL)
cv2.imshow('original img',img)
print("size of img: ", img4.shape)
cv2.namedWindow('resized img', cv2.WINDOW_NORMAL)
cv2.imshow('resized img',resized_img)
```

```python
img_hsv = cv2.cvtColor(img4, cv2.COLOR_RGB2HSV)
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR))
plt.show()
img_hsv[:, :, 0,] =  img_hsv[:, :, 0,]+5
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2BGR))
```
