```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_excel('uncleaned bike sales data.xlsx', sheet_name= 0)
df
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df.fillna(method = 'bfill')
```

```python
df.isnull().sum()
```

```python
df.Age_Group.unique()
```

```python
def age(Customer_Age):
    if Customer_Age<25:
        return 'Youth (<25)'
    if Customer_Age<35:
        return 'Young Adults (25-34)'
    return 'Adults (35-64)'
```

```python
int(df.Customer_Age[df.Age_Group.isnull()])
```

```python
df.Age_Group[df.Age_Group.isnull()] = age(int(df.Customer_Age[df.Age_Group.isnull()]))
```

```python
# df['Age_Group'] = df['Customer_Age'].apply(age)
```

```python
df = df.drop_duplicates(subset= ['Sales_Order #'], keep = 'last')
```

```python
df
```

```python
cat = df.columns[6:13]
cat
```

```python
df1 = df.copy()
classes = dict()
le = LabelEncoder()
for col in cat:
    df1[col]= le.fit_transform(df1[col])
    classes[col] = le.classes_
```
