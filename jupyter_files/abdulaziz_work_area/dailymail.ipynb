```python
import pandas as pd
import re
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from collections import Counter

# Load the CSV file
file_path = 'cnn_test.csv'
df = pd.read_csv(file_path)

# Function to perform simple tokenization using regular expressions
def simple_tokenize(text):
    # Convert text to lowercase and split into words
    return re.findall(r'\b\w+\b', text.lower())

# Function to calculate BLEU score manually without using Fraction
def modified_precision(reference, hypothesis, n):
    ref_counts = Counter([tuple(reference[i:i+n]) for i in range(len(reference)-n+1)])
    hyp_counts = Counter([tuple(hypothesis[i:i+n]) for i in range(len(hypothesis)-n+1)])
    
    overlap = {ngram: min(count, hyp_counts[ngram]) for ngram, count in ref_counts.items()}
    return sum(overlap.values()), max(1, sum(hyp_counts.values()))

def custom_bleu_score(reference, hypothesis, weights):
    p_numerators = [0] * len(weights)
    p_denominators = [0] * len(weights)

    for i, weight in enumerate(weights, start=1):
        p_numerators[i-1], p_denominators[i-1] = modified_precision(reference, hypothesis, i)

    p_n = [p_numerators[i] / p_denominators[i] if p_denominators[i] > 0 else 0 for i in range(len(weights))]
    p_log_sum = sum(weight * (0 if p == 0 else -1) for weight, p in zip(weights, p_n))
    
    return 1.0 if p_log_sum == 0 else p_log_sum

# Calculate BLEU scores using custom tokenization
smoothing_function = SmoothingFunction().method1
bleu_scores = []

for index, row in df.iterrows():
    article_tokens = simple_tokenize(row['article'])
    highlights_tokens = simple_tokenize(row['highlights'])
    # Use custom_bleu_score instead of sentence_bleu
    score = custom_bleu_score(article_tokens, highlights_tokens, [0.25, 0.25, 0.25, 0.25])
    bleu_scores.append(score)

# Add the BLEU scores to the dataframe
df['bleu_score'] = bleu_scores

# Filter the dataframe to show only the first 100 rows
df_first_100 = df[['id', 'bleu_score']].head(100)

# Display the BLEU scores for the first 100 entries
import ace_tools as tools
tools.display_dataframe_to_user(name="BLEU Scores for the First 100 Articles and Highlights", dataframe=df_first_100)
```

```python
import pandas as pd
import re
from collections import Counter
from math import exp, log

# Load the CSV file
file_path = 'cnn_test.csv'
df = pd.read_csv(file_path)

# Function to perform simple tokenization using regular expressions
def simple_tokenize(text):
    # Convert text to lowercase and split into words
    return re.findall(r'\b\w+\b', text.lower())

# Function to calculate BLEU score manually
def compute_bleu(reference, hypothesis, weights=(0.25, 0.25, 0.25, 0.25)):
    # Calculate n-gram precision
    def n_gram_precision(reference, hypothesis, n):
        ref_counts = Counter([tuple(reference[i:i+n]) for i in range(len(reference)-n+1)])
        hyp_counts = Counter([tuple(hypothesis[i:i+n]) for i in range(len(hypothesis)-n+1)])
        
        overlap = sum((min(count, hyp_counts[ngram]) for ngram, count in ref_counts.items()))
        total = sum(hyp_counts.values())
        return overlap / total if total > 0 else 0

    # Calculate brevity penalty
    def brevity_penalty(reference, hypothesis):
        ref_length = len(reference)
        hyp_length = len(hypothesis)
        if hyp_length == 0:
            return 0
        elif hyp_length > ref_length:
            return 1
        else:
            return exp(1 - ref_length / hyp_length)

    # Calculate BLEU score
    p_ns = [n_gram_precision(reference, hypothesis, n) for n in range(1, 5)]
    if all(p == 0 for p in p_ns):
        return 0
    score = exp(sum(w * log(p) if p > 0 else 0 for w, p in zip(weights, p_ns)))
    bp = brevity_penalty(reference, hypothesis)
    return score * bp

# Calculate BLEU scores
bleu_scores = []

for index, row in df.iterrows():
    article_tokens = simple_tokenize(row['article'])
    highlights_tokens = simple_tokenize(row['highlights'])
    score = compute_bleu(article_tokens, highlights_tokens)
    bleu_scores.append(score)

# Add the BLEU scores to the dataframe
df['bleu_score'] = bleu_scores

# Filter the dataframe to show only the first 100 rows with articles
df_first_100 = df[['article', 'bleu_score']].head(100)

# Display the BLEU scores for the first 100 entries with articles
print(df_first_100)
```
