```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load a global dataset (e.g., Natural Earth shapefiles)
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Filter the dataset to include only Oman
oman = world[world['name'] == 'Oman']

# Display the first few rows of the Oman GeoDataFrame
print(oman)

# Plot the map of Oman
oman.plot()
plt.title('Map of Oman')
plt.show()
```
