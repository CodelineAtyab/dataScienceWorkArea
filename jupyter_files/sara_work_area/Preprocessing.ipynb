```python

# Sample data
data = {
    'Memory': ['256GB SSD', '1TB HDD', '512GB SSD', '2TB HDD', '128GB SSD + 1TB HDD']
}
df = pd.DataFrame(data)

# Function to parse the memory column
def parse_memory(memory):
    import re
    memory = memory.replace(' ', '').split('+')
    total_memory = 0
    memory_types = []
    
    for mem in memory:
        size, unit, type_ = re.findall(r'(\d+)([A-Za-z]+)(.*)', mem)[0]
        size = int(size)
        
        # Convert size to GB
        if unit == 'TB':
            size *= 1000
        
        total_memory += size
        memory_types.append(type_)
    
    return total_memory, ' + '.join(memory_types)

# Apply the function to the memory column
df[['Memory_Size(GB)', 'Memory_Type']] = df['Memory'].apply(lambda x: pd.Series(parse_memory(x)))

print(df)
```

```python
print(df.head())
# Extract Memory Type (SSD/HDD)
df['Memory_Type'] = df['Memory'].apply(lambda x: 'SSD' if 'SSD' in x else 'HDD')

# Extract Numeric Storage Capacity
df['Memory_Capacity_GB'] = df['Memory'].str.extract(r'(\d+)GB').astype(float)

# Display the updated DataFrame with the two new columns
print(df[['Memory_Type', 'Memory_Capacity_GB']])
```

```python
df['Memory_Type'] = df['Memory'].apply(lambda x: 'SSD' if 'SSD' in x else 'HDD')

def extract_capacity(memory):
    if 'TB' or 'GB' in memory:
        return memory.split()[0]
    else:
        return None

df['Memory_Capacity'] = df['Memory'].apply(extract_capacity)

# Display the updated DataFrame with the two new columns
print(df[['Memory_Type', 'Memory_Capacity']])
```

```python
# Extract CPU Model and Clock Speed
df['CPUModel'] = df['Cpu'].apply(lambda x: ' '.join(x.split(' ')[0:3]))  # Extracts the CPU model
df['CPUClockSpeed'] = df['Cpu'].apply(lambda x: x.split()[-1])  # Extracts the CPU clock speed

# Drop the original 'CPU' column if needed
# df.drop(columns=['CPU'], inplace=True)

# Display the updated DataFrame with new CPU_Model and CPU_Clock_Speed columns
print(df[['Cpu', 'CPUModel', 'CPUClockSpeed']])
```
