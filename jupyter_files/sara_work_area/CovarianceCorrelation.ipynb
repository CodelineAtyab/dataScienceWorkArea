```python
df.Age.hist()
plt.ylabel('Frequency')
plt.xlabel('Age')
plt.show()
```

```python
df.Age.hist(color='purple', bins=10, figsize=(10, 5))
plt.show()
```

```python
# Define age bins and labels
bins = [60, 70, 80, 90, 100]
labels = ['61-70', '71-80', '81-90', '91-100']

# Categorize ages into bins
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# Count the occurrences of each age group
age_group_counts = df['Age Group'].value_counts().sort_index()

# Plot a pie chart
age_group_counts.plot.pie(
    autopct='%1.1f%%',  # Display the percentage value on the pie chart
    startangle=90,      # Start the pie chart at 90 degrees
    counterclock=False, # Draw the pie chart in a clockwise direction
    colors=plt.cm.Pastel2(range(len(labels)))  # Custom colors for the slices
)

plt.title('Age Group Distribution')
plt.show()
```
