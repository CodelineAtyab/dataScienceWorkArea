```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
img = cv2.imread('../../dataset/venuss-girdle.jpg', cv2.IMREAD_COLOR)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
img
```

```python
# Convert to grayscale
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Save or display the grayscale image
plt.imshow(gray_image, cmap='gray')
```

```python
img3 = img*10
img3 = img/10
plt.imshow(img3)
```

```python
gray = gray_image.T
plt.imshow(gray, cmap='gray')
```

```python
# resize image
resized_image = cv2.resize(img, (100, 50) ,interpolation=cv2.INTER_CUBIC)
plt.imshow(resized_image) 
```

```python
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
img_hsv[:,2,:] = img_hsv[:,2,:]*2
plt.imshow(cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB))
```
