```python
# Split the DataFrame based on the 'Company' column
company_dfs = {company: data for company, data in df.groupby('Company')}

# Example: Accessing the DataFrame for HP
hp_df = company_dfs.get('HP')

# Display the DataFrame for HP
hp_df 
```

```python
# Handle different formats in the 'Memory' column
def split_memory(memory):
    parts = memory.split(' ', 1)
    if len(parts) == 2:
        return parts
    elif len(parts) == 1:
        return [parts[0], 'Unknown']  # Handle cases where only GB is present
    else:
        return ['Unknown', 'Unknown']  # Handle unexpected cases

memory_split = df['Memory'].apply(split_memory)
memory_split_df = pd.DataFrame(memory_split.tolist(), columns=['GB', 'Memory Type'])

# Concatenate the new features with the original DataFrame
df = pd.concat([df, memory_split_df], axis=1).drop(columns=['Memory'])
print("\nDataFrame after splitting 'Memory' column:\n", df)

```

```python
# Function to split columns
def split_column(data, column_name, delimiter=' ', num_parts=2, default_fill='Unknown'):
    split_data = data[column_name].str.split(delimiter, n=num_parts-1, expand=True)
    # Handle cases where the split doesn't produce enough parts
    split_data = split_data.fillna(default_fill)
    # Rename columns if necessary
    split_data.columns = [f"{column_name}_{i+1}" for i in range(split_data.shape[1])]
    return split_data

# Example: Splitting 'Cpu' column into multiple features
cpu_split = split_column(df, 'Cpu', delimiter=' ', num_parts=3, default_fill='Unknown')

# Example: Creating dummy variables for categorical columns
categorical_columns = ['Company', 'TypeName', 'OpSys', 'Memory Type']
for col in categorical_columns:
    df = pd.concat([df, pd.get_dummies(df[col], prefix=col)], axis=1)

# Drop original categorical columns
df = df.drop(columns=categorical_columns)

# Concatenate the new features with the original DataFrame
df = pd.concat([df, cpu_split], axis=1)

# Drop original columns that have been transformed
df = df.drop(columns=['Cpu', 'ScreenResolution'])

# Display the updated DataFrame
df.head()
```
